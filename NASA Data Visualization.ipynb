```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "C:/Users/dell/Downloads/agricuture/Location hierarchy.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
print(df.head())

# Count the number of districts in each state
district_counts = df['State Name'].value_counts()

# Create a bar chart for the number of districts in each state
plt.figure(figsize=(10, 6))
plt.bar(district_counts.index, district_counts.values)
plt.xlabel('State')
plt.ylabel('Number of Districts')
plt.title('Number of Districts by State')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()

# Count the number of divisions in each state
division_counts = df['State Name'].value_counts()

# Create a bar chart for the number of divisions in each state
plt.figure(figsize=(10, 6))
plt.bar(division_counts.index, division_counts.values)
plt.xlabel('State')
plt.ylabel('Number of Divisions')
plt.title('Number of Divisions by State')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()

# Create a pie chart for the top 10 states by number of districts
top_states = district_counts.head(10)
plt.figure(figsize=(8, 8))
plt.pie(top_states.values, labels=top_states.index, autopct='%1.1f%%')
plt.title('Top 10 States by Number of Districts')
plt.show()

# Heatmap showing the number of blocks in top districts
top_districts = df['District Name'].value_counts().index[:10]
pivot_df = df[df['District Name'].isin(top_districts)].pivot_table(
    index='District Name', columns='State Name', aggfunc='size', fill_value=0
)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df, cmap='Blues', annot=True, fmt='d')
plt.title('Number of Blocks in Top Districts')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
file_path = "C:/Users/dell/Downloads/agricuture/IMD Agromet advisory locations.xlsx"
df = pd.read_excel(file_path)

# Display the first few rows to understand the structure
print(df.head())
print(df.columns)  # Display all column names for exploration

# Assuming we have columns like 'State', 'District', and 'Station Name'

# Count the number of stations in each state
state_counts = df['State'].value_counts()

# Create a bar chart for the number of stations in each state
plt.figure(figsize=(12, 6))
plt.bar(state_counts.index, state_counts.values, color='skyblue')
plt.xlabel('State')
plt.ylabel('Number of Stations')
plt.title('Number of Agromet Stations by State')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Count the number of stations in each district
district_counts = df['District'].value_counts()

# Create a bar chart for the number of stations in each district
plt.figure(figsize=(12, 6))
plt.bar(district_counts.index[:20], district_counts.values[:20], color='lightgreen')  # Top 20 districts
plt.xlabel('District')
plt.ylabel('Number of Stations')
plt.title('Top 20 Districts by Number of Agromet Stations')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Create a pie chart for the top 10 states by number of stations
top_states = state_counts.head(10)
plt.figure(figsize=(8, 8))
plt.pie(top_states.values, labels=top_states.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Top 10 States by Number of Agromet Stations')
plt.show()

# Heatmap for the distribution of stations across the top 10 districts and states
top_districts = df['District'].value_counts().index[:10]
top_states = df['State'].value_counts().index[:10]

pivot_df = df[df['District'].isin(top_districts) & df['State'].isin(top_states)].pivot_table(
    index='District', columns='State', aggfunc='size', fill_value=0
)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df, cmap='coolwarm', annot=True, fmt='d')
plt.title('Heatmap of Agromet Stations in Top 10 Districts and States')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path_1 = "C:/Users/dell/Downloads/agricuture/IMD Agromet advisory locations.xlsx"
df_1 = pd.read_excel(file_path_1)

# Data preprocessing: Remove rows where District or State is missing
df_1_clean = df_1.dropna(subset=['State', 'District'])

# 1. Bar chart: Count of districts per state
state_counts = df_1_clean['State'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=state_counts.index, y=state_counts.values, palette='viridis')
plt.xlabel('State')
plt.ylabel('Number of Districts with Agromet Advisory')
plt.title('Number of Districts per State with IMD Agromet Advisory')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()

# 2. Pie chart: Top 10 states by number of districts with Agromet advisory
top_states = state_counts.head(10)

plt.figure(figsize=(8, 8))
plt.pie(top_states.values, labels=top_states.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))
plt.title('Top 10 States by Number of Districts with IMD Agromet Advisory')
plt.show()

# 3. Heatmap: Number of districts in each state
state_district_pivot = pd.crosstab(index=df_1_clean['District'], columns=df_1_clean['State'])

plt.figure(figsize=(15, 10))
sns.heatmap(state_district_pivot, cmap='coolwarm', cbar=True, linewidths=.5)
plt.title('Heatmap of Districts per State with IMD Agromet Advisory')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'C:/Users/dell/Downloads/agricuture/Agmark crops.xlsx'

try:
    # Load specific sheet
    df = pd.read_excel(file_path, sheet_name='Gramhal - Agmark Crops')

    # Ensure the column exists
    if 'Crop Types Type Meanings' not in df.columns:
        raise KeyError(f"Column 'Crop Types Type Meanings' not found. Available columns: {df.columns}")

    # Count the occurrences of each crop type
    crop_type_counts = df['Crop Types Type Meanings'].value_counts()

    # Set the visualization style
    sns.set(style="whitegrid")

    # Create a bar chart
    plt.figure(figsize=(10, 6))
    sns.barplot(x=crop_type_counts.index, y=crop_type_counts.values, palette='viridis')
    plt.title('Number of Crops by Type', fontsize=16)
    plt.xlabel('Crop Types', fontsize=12)
    plt.ylabel('Number of Crops', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()  # To avoid cutting off labels
    plt.show()

    # Create a pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(crop_type_counts, labels=crop_type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(crop_type_counts)))
    plt.title('Distribution of Crop Types', fontsize=16)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

except FileNotFoundError:
    print(f"File not found at path: {file_path}")
except KeyError as e:
    print(e)
except Exception as e:
    print(f"An error occurred: {e}")
```
